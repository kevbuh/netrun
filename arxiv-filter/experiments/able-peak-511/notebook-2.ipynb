```
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, linewidth=2, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Basic Sine Wave Plot')
plt.grid(True, alpha=0.3)
```
